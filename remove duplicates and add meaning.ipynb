```python
import json
import requests
import pandas as pd
from tqdm import tqdm
```

# Get list of words

```python
words_list = list(pd.read_csv("words_list.csv", header=None)[0])
words_list = list(set(words_list)) # unique
words_list.sort() # ascending order cause why not
words_list
```

# Add meaning

```python
def getUrlResponse(word):
    
    base_url = 'https://api.dictionaryapi.dev/api/v2/entries/en/'
    
    resp = requests.get(url=base_url+word).json()

    try:
        if resp['title'] == 'No Definitions Found':
            print("Not a word")
    except:
        pass
    
    return resp

getUrlResponse('acclaim')
```

```python
def get_meanings_list_from_single_response(single_response):

    list_of_meaning_dict = single_response['meanings']
    list_of_definitions = []
    
    for meaning_dict in list_of_meaning_dict:
        partOfSpeech = meaning_dict['partOfSpeech']
        for definition_dict in meaning_dict['definitions']:
            list_of_definitions.append(partOfSpeech +': ' + definition_dict['definition'])
        
    return list_of_definitions
```

```python
words = []
defns = []

for word in tqdm(words_list):
    
    responses = getUrlResponse(word)
    
    list_of_defn = []
    for response in responses:
        list_of_defn.extend(get_meanings_list_from_single_response(response))
    
    words.append(word)
    defns.append(list_of_defn)
```
